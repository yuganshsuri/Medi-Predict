```python
disease_dict={'Healthy':0, 'Pre-Diabetes':1, 'Hypertension':2, 'Heart Disease':3, 'Diabetes':4}
smoking_dict={'No':0, 'Yes':1}
alchol_dict={'No':0, 'Yes':1}
family_dict={'No':0, 'Yes':1}
gender_dict={'Male':0, 'Female':1}
activity_dict={'Medium':0, 'High':1, 'Low':2}
```

```python
df["Disease"]=df["Disease"].map(disease_dict)
df["Smoking"]=df["Smoking"].map(smoking_dict)
df["Gender"]=df["Gender"].map(gender_dict)
df["FamilyHistory"]=df["FamilyHistory"].map(family_dict)
df["PhysicalActivity"]=df["PhysicalActivity"].map(activity_dict)
df["Alcohol"]=df["Alcohol"].map(alchol_dict)
```
